# Karmaşıklık Matrisi (Confusion Matrix)

**Basitçe bir eksende tahminlerin, diğer eksende asıl değerlerin bulunduğu bir matris**

**Diagondaki değerler bizim için doğru değerler**



**Accuracy = Bizim doğru değerlerimizin tüm değerlere oranı**

**Missclassification rate = Yanlış değerlerin tüm değerlere oranı ---- (1-acc)**

**Acc değeri, bizim için sınıflandırma doğruluğunu ölçen en önemli kriter ancak bazı durumlarda veriler dengesiz dağılmış olabilir. Algoritmamız başarısızsa, her önüne gelene pozitif diyorsa mesela algoritmamız zeki değilse, verileri öğrenmiyorsa vb. %99 başarı elde edip %1 hata elde edersek %99 başarı bile bizim için başarı değildir.**

****

**True positive kısmında toplam 105 tane yes var. Bunlardan kaç tanesi gerçekten yes, onu bulmamız gerek. False positive'de ise no'ların ne kadar doğru sınıflandırıldığını buluyoruz. 60 taneden 10 tanesi aslında no iken yes olarak görülmüş.**

**Prevalence değeri bizim ne kadar dengeli sınıfımızın olduğunu söyler. 2 sınıfımız var ise %50 değeri bizim dengeli verilerimizin olduğunu söyler. %50'den artıp azaldıkça dengesizlik artar.**



# False Pozitif ve False Negatif Kavramları

**Makine öğrenmesinde her şey bir hipotez ile başlıyor ve bu kabulün doğru ya da yanlış olması ile devam ediyor. Doğru karar verirsek karar ya doğru sonuçlanır ya da yanlış sonuçlanır. Yanlış sonuçlanırsa tip 1 hatası alınır, doğru sonuçlanırsa doğru karar verilmiş olur. Eğer yanlış olduğuna karar verirsek ve doğru çıkarsa tip 2 hatası alınır.**

****

**SVM'de hata kavramına bakarsak: üst kısım pozitif, alt kısım negatif diyelim. İçi dolu olanlar train, içi boş olanlar test verileri. Yani yuvarlak olanlar pozitif, kare olanlar negatif anlamına geliyor. Biz marjin alanında baktığımız zaman negatiflerin arasında yuvarlak bir değer görürsek bu bir hatadır, yani false negatif olur. Aynı şekilde pozitiflerin arasındaki marjinde bir kare şekil varsa bunlar false pozitif olduğu anlamına gelir.**



**Logistik regresyonda hata için: Sigmoid bir fonksiyondur(çünkü içindeki fonksiyon s şekline doğru gidiyor) Gerçekte veri herhangi bir yerde bulunabilir. Biz bu algoritma ile verileri 0'a ya da 1'e yuvarlıyorduk. Tam ortadaki geçişin sol tarafında bir değer üst kısımda kalırsa bu bir hatadır (false negative). Aynı şekilde geçiş yerinin sağında bir değer alt kısımda kalırsa bu da false positive olur. Aslında pozitif sınıflandırmada yer alması gerekirken aşağıda olduğu için yanlış bir sınıflandırma olur.**



**Karar ağacında ise erkeğe pozitif, kadına negatif dersek kadın için doğru bulduğu değerler true negative, yanlış bulduğu deperler false negative. Erkek için doğru bulduğu değerler true positive, yanlış bulduğu değerler false positive olur.**


# Doğruluk(Accuracy) Paradoksu

**ZeroR algoritmasını ele alırsak, prevalence değerine göre sınıflandırma yapıyoruz. Yani toplam positive değerleri = gerçek+tahmin, toplam negative değerleri = gerçek + toplam**



**Hangi sınıftan sana daha çok örnek geldiyse onu sınıflandırma sonucu olarak kullan. Hiçbir şeye bakma, hiçbir şeyi öğrenme. Sadece sana verilen sonuçları kullan! Her gelene direkt c1 diyecek, c2'de tahmin yapmayacak. Önceden daha az doğruluk oranı varken artık doğruluk oranımız artar ve %99 oluyor. Ancak hiçbir şey öğrenmiyor ve yaptığı bir sınıflandırma yok!**

**Sadece accuracy'e bakarak sınıflandırma algoritmalarını değerlendirmek doğru olmayabilir. Buradan bunu anlıyoruz. Bir sınıflandırma algoritmasını başarılı kabul edebilmemiz için bir alt çizgiye ihtiyacımız var. Eğer değerler bu çizginin altına düşüyorsa o algoritma çok da başarılı, çok da öğrenen bir algoritma değil. ZeroR bile ondan daha başarılı duruma düşüyorsa o algotirmada bir sıkıntı vardır. True negative, false pozitive'leri bulmak da bir başarı olabilir.**

**Bir algoritma zeroR algoritmasını geçemiyorsa o algoritma çok başarısızdır diyebiliriz. Yazdığımız algoritma çok başarılı bir algoritma olmayabilir. Bunu da göz önünde bulundurmamız gerek**

# ROC (Receiver Operating Characteristic) Eğrisi

**True positive ve false positive değerlerini hesaplıyoruz. Verileri bir ROC eğrisi üzerine koyuyoruz. Yani, bir eksende fp, bir eksende tp değerlerimiz var.**



**Eksendeki kırmızı noktalar bizim sınıflandırma algoritmalarımız. Fp'nin %100 olması demek, tüm değerlerin false positive'de olması yani true negativede hiç değer olmaması demek. Yani %100 hatamız var.**

**Tp'nin %100 olması demek bizim için doğru olması anlamına geliyor. Ancak bizim hem %100 fp hem de %100 tp'ye sahip olmamız. no olan sonuçlarda %100 hata, yes olan sonuçlarda %100 doğruluk anlamına geliyor.**

**Ancak sol üst köşedeki kırmızı nokta, fp'nin %0, tp'nin %100 olduğu anlamına geliyor. Yani çok güzel bir sonuca ulaşmışız demektir. Çizginin üzerindeki noktalar bizim hatalı noktalarımız. Bir tarafta kazanırken bir tarafta kaybettiğimiz noktalar.**

**Beklediğimiz kadarıyla çizginin altına inemeyiz. Eğer fp %100, tp %0 olursa bizim aslında başarılı bir algoritmamız var. Tam tersi olarak aldığımızda tp %100, fp %0 olur. Bir şeyler öğreniyor ancak tersten öğretiyor demektir.**

****

**Diyelim 5 farklı algoritma aldık ve elimizde böyle bir şekil var. Hangisini alacağımıza nasıl karar vereceğiz?**

**Böyle bir durumda doğrusal olarak birleştirmeye kalkarsak bazı noktalar bu alanın içinde kalır (C3 gibi). Yani aslında bize faydalı olmadığını anlarız. C2 ve C4, C3'ü domine ediyor.**

****

**Daha sonra prevalence değerine göre acc doğruları çiziliyor. Eğer prev değerimiz 0.5 ise yani dengeli ise 45 derecelik acc doğruları çizilir(çizgiden 2 kare aşağı ve 2 kare sola sayabiliriz) (sol şekil). Ancak prev değerimiz 0.5'ten farklı, yani 0.33 veya 0.66 ise yaklaşık, eğrilerimiz 2'ye 4 olur (sağ şekil). Diyelim %60 başarılı bir algoritma var (%50 diye ok çıkarılmış çizginin üst tarafı) o çizgi üzerinde gidildikçe tpr'nin %20 olduğu kısımda da o çizginin devamının en üste gittiği tarafta da bizim başarımız %60 olur çünkü true positive değeri azalırsa true negative değeri artar yani hep dengeli devam eder.**



**Diyelim prev değeri 0.5, böyle bir dünyada acc'leri çizdiğimiz zaman c2'de acc bize %80 başarı veriyor. Yani acc tablosuna göre mavi çizgileri çiziyoruz. Bu acc doğrusuna göre en fazla başarı C2'de. Ortadaki şekil C4'ü, sondaki şekil ise C1'i gösteriyor. C4 acc doğrusuna baktığımız zaman, neg/pos oranı 1/4 olduğunda C4, %82 civarında başarı veriyor yani hepsinden fazla. Orana göre değişimler gözüküyor.**

# AUC (Area Under Curve) Değeri

**Elimizdeki veri kümesi arttıkça acc değerimiz önce artar çünkü daha fazla veriden öğrenir. Belirli bir noktadan sonra sabit devam eder.**

**AUC değeri, ROC ile doğrular çizdiğimizde alanın içerisinde kalan kısımdır.**



**AUC değeri ise bu eğrinin altında kalan kısım. Dörtgen kısmı ile en alttaki üçgen kısmınının arasında gidip gelen değer.**

**AUC değeri ne kadar büyükse bizim ulaşabileceğimiz acc değeri o kadar yüksek olur**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_curve
```

```python
datas = pd.read_csv("veriler.csv")
datas
```

```python
bky = datas.iloc[:, 1:4]
cins = datas.iloc[:, -1]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(bky, cins, test_size=0.33, random_state=0)
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
#Rassal Ağaçlar ile roc, fpr, tpr değerlerini bulma
rf = RandomForestClassifier(n_estimators=10, criterion='entropy')
rf.fit(xtrain, ytrain)
r_pred = rf.predict_proba(xtest)
print(ytest)
print(r_pred)

#şu an predict_proba ile ihtimalleri yazdırdık. Yani erkek ve kadın olma ihtimallerine bakıyor
#hangisi daha yüksekse ona göre erkek ya da kadın diye sınıflandırma yapıyor.
#her bir sınıfın alabileceği olasılıklar. 0. kolon erkek, 1. kolon kadın olarak alıyor.
#yani erkek kolonuna bakarsak ilk kişinin %60 erkek olduğuna güvenebiliriz. 2. kişinin %20, 3.kişinin %20 vb. güvenebiliriz anlamına gelir!
```

```python
#roc değerini, fpr'yi ve tpr'yi hesaplayalım. Bize sırasıyla fpr, tpr, thresholds döndürecek
#roc_curve fonksiyonu 3 tane parametre alacak. y_true, y_score, pos_label istiyor.
#ytest bizim gerçek değerlerimiz. y_score, bizim probalarımız. pos_label ise pozitiflerin labellarını istiyor.
#yani pos_label bizim ytestimizdeki almak istediklerimiz. Diyelim erkekleri almak istiyoruz
fpr, tpr, thold = roc_curve(ytest, r_pred[:,0], pos_label='e')
print("tpr = ", tpr)
print("fpr = ", fpr)
print("thold = ", thold)#bu değer eşik değerdir. Bulduğumuz değer bu eşik değerin üstündeyse kadın, altındaysa erkek diye sınıflandırılabilir.
                        
```
