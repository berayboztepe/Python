# DeÄŸiÅŸken SeÃ§imi

**Birden fazla deÄŸiÅŸken olduÄŸunda acaba her deÄŸiÅŸken, baÄŸÄ±mlÄ± deÄŸiÅŸkeni aynÄ± oranda mÄ± etkiliyor? hatta bazÄ± deÄŸiÅŸkenler hiÃ§ etkilemiyor bile olabilir. BazÄ± deÄŸiÅŸkenler sisteme olumsuz etki de yapabilir.**

### 1-) BÃ¼tÃ¼n DeÄŸiÅŸkenleri Dahil Etmek: 
**Elimizdeki tÃ¼m deÄŸiÅŸkenleri dahil edip hiÃ§birini elememek. BazÄ± durumlarda bu yÃ¶ntemi kullanmak zorunda kalabiliriz(sistemi olumsuz etkilese bile). Mesela hali hazÄ±rda bulunan bir sistemin baÅŸarÄ±sÄ±nÄ± Ã¶lÃ§mek gibi(Ã¶zellikleri Ã¶nceden seÃ§ilmiÅŸ) veya KeÅŸif iÃ§in denenebilir(diÄŸer yÃ¶ntemleri kullanmadan bir Ã¶n fikir elde etmek iÃ§in)**

### 2-) Geriye DoÄŸru Eleme(Backward Elimination):
**Ã‡ok kullanÄ±lan bir yÃ¶ntemdir. Ã–nce bÃ¼tÃ¼n deÄŸiÅŸkenler dahil edilir, sonra teker teker elenerek ilerlenir. 
En yÃ¼ksek p-value deÄŸerine sahip olan deÄŸiÅŸken ele alÄ±nÄ±r. P>0.05 ise bir sonraki adÄ±ma geÃ§ilir, deÄŸilse makine Ã¶ÄŸrenmesi sonlandÄ±rÄ±lÄ±r.
Daha sonra, en yÃ¼ksek deÄŸere sahip p-value sistemden kaldÄ±rÄ±lÄ±r ve yeni deÄŸiÅŸen seÃ§ilir. Yeniden p>0.05 olup olmadÄ±ÄŸÄ±na bakÄ±lÄ±r... BÃ¶yle devam eder.**

### 3-) Ä°leriye DoÄŸru Eleme(Forward Selection):
**Yeniden dÃ¶ngÃ¼ ile yapÄ±lÄ±r(2. sinde olduÄŸu gibi) En dÃ¼ÅŸÃ¼k deÄŸere sahip olan p-value deÄŸiÅŸkeni alÄ±nÄ±r. 
P<0.05 saÄŸlandÄ±ÄŸÄ± sÃ¼rece yeni deÄŸiÅŸken eklenmeye devam edilir. Ne zaman ki p<0.05 olursa makine Ã¶ÄŸrenimi sonlanÄ±r.**

### 4-) Ã‡ift YÃ¶nlÃ¼ Eleme(Bidirectional Elemination)
**2 ve 3'Ã¼n birleÅŸimi. AmacÄ±mÄ±z hem ileri hem geri ilerlemek. BazÄ± durumlarda birden fazla Significant level(0.05) belirlenebilir: Biri ileriye giderken, biri geri giderken.
Yine en dÃ¼ÅŸÃ¼k p-value deÄŸeri ele alÄ±nÄ±r. Bu deÄŸiÅŸken sabit tutularak diÄŸer tÃ¼m deÄŸiÅŸkenler sisteme dahil edilir. 
SL deÄŸiÅŸkeninin altÄ±nda kalan deÄŸiÅŸkenler sistemde kalÄ±r, eski deÄŸiÅŸkenlerden hiÃ§biri sistemden Ã§Ä±karÄ±lmaz.**

### 5-) Skor KarÅŸÄ±laÅŸtÄ±rmasÄ±(Score Comparison):
**P-value, Sl falan belirlemeyip  tÃ¼m kurallarÄ± kendimiz koyarÄ±z. Bir baÅŸarÄ± kriteri belirleyeceÄŸiz. OlasÄ± regresyon yÃ¶ntemleri inÅŸa edilir. Belirlenen kriter saÄŸlanana kadar inÅŸa devam eder.**

```python
data = pd.read_csv("veriler.csv")
data
```

```python
#cinsiyet ve Ã¼lke kolonlarÄ±nÄ± numerik deÄŸere Ã§evireceÄŸiz.
ulke=data.iloc[:,0:1].values
le = preprocessing.LabelEncoder()
ulke[:,0] = le.fit_transform(data.iloc[:,0])

ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke).toarray()

cinsiyet = data.iloc[:, -1:]
ohe = preprocessing.OneHotEncoder()
cinsiyet = ohe.fit_transform(cinsiyet).toarray()
```

```python
#Åžimdi dataframeleri birleÅŸtiriyoruz
df = pd.DataFrame(data=ulke, index=range(22), columns=['fr', 'tr', 'usa'])

Yas = data.iloc[:, 1:4]
df1 = pd.DataFrame(data=Yas, index=range(22), columns=['boy', 'kilo', 'yas'])

#ancak cinsiyet alÄ±rken dummy variablelara dikkat ediyoruz. Ondan sadece tek kolonu alacaÄŸÄ±z. Sadece erkek kolonunu aldÄ±k!
#cinsiyet = 1 ise erkek, 0 ise kÄ±z
df3 = pd.DataFrame(data=cinsiyet[:, 0:1], index=range(22), columns=['cinsiyet'])

sonuc1df = pd.concat([df, df1], axis=1)

sonuc3df = pd.concat([sonuc1df, df3], axis=1)#Cinsiyet One-Hot Endocing'i iÃ§in bir df
sonuc3df
```

```python
#yeniden train ve test bloklarÄ±mÄ±zÄ± oluÅŸturuyoruz.
#baÄŸÄ±mlÄ± deÄŸiÅŸken cinsiyet kolonu
x_train, x_test, y_train, y_test = train_test_split(sonuc1df, df3, test_size=0.33, random_state=0)

print(x_train)
print("\n\n", y_train)
print("\n\n",x_test)
print("\n\n",y_test)
```

```python
#Åžimdi ise Multiple Linear Regression iÅŸlemlerine geÃ§iyoruz.
lr = LinearRegression()
lr.fit(x_train, y_train)#6 boyutlu bir veri kullanÄ±lacak
```

```python
#tahmin iÅŸlemine geÃ§elim
#gÃ¶rÃ¼ldÃ¼ÄŸÃ¼ gibi 8 Ã¶rnekte 2 hata var(kalanÄ±nÄ± doÄŸru kabul edelim)
y_predict = lr.predict(x_test)
print(y_predict)
print(y_test)
```

```python
#Åžimdi ise diÄŸer deÄŸerlere gÃ¶re boyu tahmin etmeye Ã§alÄ±ÅŸacaÄŸÄ±z. Yani boy harici x, boy y deÄŸeri olacak
#boy'u ayri bir deÄŸer olarak aldÄ±k. Boyun solunda ve saÄŸÄ±nda kalan deÄŸiÅŸkenleri aldÄ±k ve yeni bir df olarak birleÅŸtirdik
boy = sonuc3df.iloc[:, 3:4]
sol = sonuc3df.iloc[:, :3]
sag = sonuc3df.iloc[:, 4:]

newdf = pd.concat([sol, sag], axis=1)
newdf
```

```python
#yeniden train ve test kÃ¼meleri oluÅŸturuyoruz

xtrain, xtest, ytrain, ytest = train_test_split(newdf, boy, test_size=0.33, random_state=0)
print(xtrain)
print("\n\n", ytrain)
print("\n\n",xtest)
print("\n\n",ytest)
```

```python
#ÅŸimdi yeniden lineer Ã¶ÄŸretme yapÄ±yoruz
lr.fit(xtrain, ytrain)
y_boy_predict = lr.predict(xtest)
print(y_boy_predict)
print(ytest)
```
