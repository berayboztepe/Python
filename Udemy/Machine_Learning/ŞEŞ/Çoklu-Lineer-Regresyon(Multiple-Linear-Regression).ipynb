# Dummy Variable(Kukla Değişken)
**Bir değişkeni ifade eden başka değişken diyebiliriz. Örneğin cinsiyet kategorik bir veriydi. Ancak biz OHE yapıp onu numerik veri haline getirdik. Bu kolon kukla değişken olur**

**Ancak bu işlem her zaman iyi sonuç vermez. Bazı makine öğrenmesi algoritmaları bu işleme dirençliyken bazı makine öğrenmesi algoritmaları dirençli değildir. Mesela bir kolon için 100 farklı kolon eklediğimizde, önceden varolan eski kolonlar işlevini yitirmeye başlar çünkü algoritma kendini bu yeni eklenen 100 kolona doğru çeker.**

**Dummy Variable oluşturduktan sonra mutlaka yeni oluşturulan kolonlarla bu kolonların oluşturulduğu kolon aynı anda alınmamalıdır. # P Value(Probility Value)

**h0(sıfır hipotezi) = başlangıçta ortaya atılan hipotez olarak geçer. "Her kutuda 100 tane kurabiye vardır."**

**h1(ters hipotez) = Başlangıçta verdiğimiz hipotezin tersi. Hipotezi ne zaman çürütebiliriz? 1 tane hatalı değer(outlier) bu hipotezi çürütemeyebilir.**

**P value = Kaç tane aksi örnek bulursam bu hipotezi çürütebilirim? bunun cevabını verir.**

**p value = genelde %5 alınır. p değeri %5'in altındaysa h0'ın hatalı olma ihtimali artar. Dolayısıyla h1'in doğru olma değeri artar.(h0 ve h1 birbirinin tersi.**

# Değişken Seçimi

**Birden fazla değişken olduğunda acaba her değişken, bağımlı değişkeni aynı oranda mı etkiliyor? hatta bazı değişkenler hiç etkilemiyor bile olabilir. Bazı değişkenler sisteme olumsuz etki de yapabilir.**

### 1-) Bütün Değişkenleri Dahil Etmek: 
**Elimizdeki tüm değişkenleri dahil edip hiçbirini elememek. Bazı durumlarda bu yöntemi kullanmak zorunda kalabiliriz(sistemi olumsuz etkilese bile). Mesela hali hazırda bulunan bir sistemin başarısını ölçmek gibi(özellikleri önceden seçilmiş) veya Keşif için denenebilir(diğer yöntemleri kullanmadan bir ön fikir elde etmek için)**

### 2-) Geriye Doğru Eleme(Backward Elimination):
**Çok kullanılan bir yöntemdir. Önce bütün değişkenler dahil edilir, sonra teker teker elenerek ilerlenir. 
En yüksek p-value değerine sahip olan değişken ele alınır. P>0.05 ise bir sonraki adıma geçilir, değilse makine öğrenmesi sonlandırılır.
Daha sonra, en yüksek değere sahip p-value sistemden kaldırılır ve yeni değişen seçilir. Yeniden p>0.05 olup olmadığına bakılır... Böyle devam eder.**

### 3-) İleriye Doğru Eleme(Forward Selection):
**Yeniden döngü ile yapılır(2. sinde olduğu gibi) En düşük değere sahip olan p-value değişkeni alınır. 
P<0.05 sağlandığı sürece yeni değişken eklenmeye devam edilir. Ne zaman ki p<0.05 olursa makine öğrenimi sonlanır.**

### 4-) Çift Yönlü Eleme(Bidirectional Elemination)
**2 ve 3'ün birleşimi. Amacımız hem ileri hem geri ilerlemek. Bazı durumlarda birden fazla Significant level(0.05) belirlenebilir: Biri ileriye giderken, biri geri giderken.
Yine en düşük p-value değeri ele alınır. Bu değişken sabit tutularak diğer tüm değişkenler sisteme dahil edilir. 
SL değişkeninin altında kalan değişkenler sistemde kalır, eski değişkenlerden hiçbiri sistemden çıkarılmaz.**

### 5-) Skor Karşılaştırması(Score Comparison):
**P-value, Sl falan belirlemeyip  tüm kuralları kendimiz koyarız. Bir başarı kriteri belirleyeceğiz. Olası regresyon yöntemleri inşa edilir. Belirlenen kriter sağlanana kadar inşa devam eder.**

```python
data = pd.read_csv("veriler.csv")
data
```

```python
#cinsiyet ve ülke kolonlarını numerik değere çevireceğiz.
ulke=data.iloc[:,0:1].values
le = preprocessing.LabelEncoder()
ulke[:,0] = le.fit_transform(data.iloc[:,0])

ohe = preprocessing.OneHotEncoder()
ulke = ohe.fit_transform(ulke).toarray()

cinsiyet = data.iloc[:, -1:]
ohe = preprocessing.OneHotEncoder()
cinsiyet = ohe.fit_transform(cinsiyet).toarray()
```

```python
#Şimdi dataframeleri birleştiriyoruz
df = pd.DataFrame(data=ulke, index=range(22), columns=['fr', 'tr', 'usa'])

Yas = data.iloc[:, 1:4]
df1 = pd.DataFrame(data=Yas, index=range(22), columns=['boy', 'kilo', 'yas'])

#ancak cinsiyet alırken dummy variablelara dikkat ediyoruz. Ondan sadece tek kolonu alacağız. Sadece erkek kolonunu aldık!
#cinsiyet = 1 ise erkek, 0 ise kız
df3 = pd.DataFrame(data=cinsiyet[:, 0:1], index=range(22), columns=['cinsiyet'])

sonuc1df = pd.concat([df, df1], axis=1)

sonuc3df = pd.concat([sonuc1df, df3], axis=1)#Cinsiyet One-Hot Endocing'i için bir df
sonuc3df
```

```python
#yeniden train ve test bloklarımızı oluşturuyoruz.
#bağımlı değişken cinsiyet kolonu
x_train, x_test, y_train, y_test = train_test_split(sonuc1df, df3, test_size=0.33, random_state=0)

print(x_train)
print("\n\n", y_train)
print("\n\n",x_test)
print("\n\n",y_test)
```

```python
#Şimdi ise Multiple Linear Regression işlemlerine geçiyoruz.
lr = LinearRegression()
lr.fit(x_train, y_train)#6 boyutlu bir veri kullanılacak
```

```python
#tahmin işlemine geçelim
#görüldüğü gibi 8 örnekte 2 hata var(kalanını doğru kabul edelim)
y_predict = lr.predict(x_test)
print(y_predict)
print(y_test)
```

```python
#Şimdi ise diğer değerlere göre boyu tahmin etmeye çalışacağız. Yani boy harici x, boy y değeri olacak
#boy'u ayri bir değer olarak aldık. Boyun solunda ve sağında kalan değişkenleri aldık ve yeni bir df olarak birleştirdik
boy = sonuc3df.iloc[:, 3:4]
sol = sonuc3df.iloc[:, :3]
sag = sonuc3df.iloc[:, 4:]

newdf = pd.concat([sol, sag], axis=1)
newdf
```

```python
#yeniden train ve test kümeleri oluşturuyoruz

xtrain, xtest, ytrain, ytest = train_test_split(newdf, boy, test_size=0.33, random_state=0)
print(xtrain)
print("\n\n", ytrain)
print("\n\n",xtest)
print("\n\n",ytest)
```

```python
#şimdi yeniden lineer öğretme yapıyoruz
lr.fit(xtrain, ytrain)
y_boy_predict = lr.predict(xtest)
print(y_boy_predict)
print(ytest)
```
